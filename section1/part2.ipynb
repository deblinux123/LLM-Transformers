```python
from transformers import pipeline

unmasker = pipeline("fill-mask", model="bert-base-uncased")
result = unmasker("This man works as a [MASK].")
print([r["token_str"] for r in result])

result = unmasker("This woman works as a [MASK].")
print([r["token_str"] for r in result])
```

هنگامی که از مدل خواسته می‌شود کلمه‌ی حذف‌شده را در این دو جمله تکمیل کند، تنها **یک پاسخ بدون سوگیری جنسیتی** (مانند *waiter/waitress*، یعنی پیشخدمت) ارائه می‌دهد. سایر پاسخ‌ها عمدتاً **مشاغلی هستند که معمولاً در جامعه با یک جنسیت خاص مرتبط دانسته می‌شوند**. حتی واژه **«فاحشه» (prostitute)** نیز در میان **پنج گزینه برتر** برای جمله‌ای که شامل واژه **«زن» (woman)** و **«کار» (work)** بود، قرار گرفت.

این موضوع در حالی رخ می‌دهد که **BERT** یکی از معدود مدل‌های مبتنی بر **Transformer** است که داده‌های آموزشی آن از خزش گسترده کل اینترنت جمع‌آوری نشده‌اند؛ بلکه با استفاده از مجموعه‌داده‌هایی که در ظاهر نسبتاً خنثی به نظر می‌رسند، یعنی **English Wikipedia** و **BookCorpus**، آموزش دیده است.

این مثال نشان می‌دهد که حتی مدل‌هایی که روی داده‌های نسبتاً کنترل‌شده آموزش دیده‌اند نیز می‌توانند **سوگیری‌های موجود در داده‌های آموزشی** را یاد بگیرند و هنگام تولید خروجی آن‌ها را بازتاب دهند.

بنابراین، هنگام استفاده از این مدل‌ها باید همیشه این نکته را در نظر داشته باشید که **مدل پایه (Base Model)** ممکن است به‌راحتی محتوایی **جنسیت‌زده (Sexist)**، **نژادپرستانه (Racist)** یا **تبعیض‌آمیز نسبت به گرایش‌های جنسی (Homophobic)** تولید کند.

همچنین باید توجه داشت که **ریزتنظیم (Fine-tuning)** مدل بر روی داده‌های اختصاصی شما، **لزوماً این سوگیری‌های ذاتی را از بین نمی‌برد**. اگرچه ریزتنظیم می‌تواند رفتار مدل را در یک حوزه خاص بهبود دهد، اما سوگیری‌هایی که در مرحله پیش‌آموزش در مدل نهادینه شده‌اند ممکن است همچنان در خروجی آن مشاهده شوند.

به همین دلیل، هنگام توسعه و استفاده از سامانه‌های مبتنی بر مدل‌های زبانی بزرگ، لازم است خروجی‌ها به‌طور مستمر **ارزیابی، پایش و در صورت امکان با استفاده از روش‌های کاهش سوگیری (Bias Mitigation)** کنترل شوند، به‌ویژه در کاربردهایی که با افراد، تصمیم‌گیری یا موضوعات حساس سروکار دارند.
